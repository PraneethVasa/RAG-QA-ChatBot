### OpenAI Embeddings

```python
import os
from dotenv import load_dotenv
load_dotenv
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model='text-embedding-3-large')
embeddings
```

```python
input_txt = "The RecursiveCharacterTextSplitter in LangChain is a powerful text-splitting method that ensures intelligent chunking of documents while maintaining context"
result = embeddings.embed_query(input_txt)
result
```

### Ollama Embeddings

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings = ( OllamaEmbeddings(model='gemma:2b') )
```

```python
embeddings
```

```python
input_txt = "The RecursiveCharacterTextSplitter in LangChain is a powerful text-splitting method that ensures intelligent chunking of documents while maintaining context"
result = embeddings.embed_query(input_txt) #embed_documents for documents or list of texts
result
```

```python
len(result)
```

```python
input_txt1 = "The RecursiveCharacterTextSplitter in LangChain is a powerful text-splitting method that ensures intelligent chunking of documents while maintaining context."

input_txt2 = "This approach allows for better retrieval-augmented generation (RAG) results by preserving meaningful sections of text and improving embeddings."

input_txt3 = "By using multiple recursive separators, it can efficiently split large documents, keeping semantic meaning intact."

input_txt4 = "This makes it ideal for applications like question-answering systems, chatbots, and search engines that rely on accurate text embeddings."

result = embeddings.embed_documents([input_txt1, input_txt2, input_txt3, input_txt4]) #embed_documents for documents or list of texts
result
```
